```python
from urllib import response
import dotenv
import os
from requests import Session
from pprint import pprint

dotenv.load_dotenv()

API_KEY = os.getenv("COINMARKETCAP_API_KEY")
API_ENDPOINT = "https://pro-api.coinmarketcap.com/v2/cryptocurrency/quotes/latest"

request_symbol = "BTC"

parameters = {
    "symbol": request_symbol,
    "convert": "USD",
}

headers = {
    "Accepts": "application/json",
    "X-CMC_PRO_API_KEY": API_KEY,
}

s = Session()
s.headers.update(headers)

response_api = s.get(API_ENDPOINT, params=parameters)
data = response_api.json().get("data")
data_symbol = next(iter(data.get(request_symbol)))

pprint(data)
```

```python
for k in data_symbol.keys():
    print(k, type(data_symbol.get(k)), sep="\t")
```
